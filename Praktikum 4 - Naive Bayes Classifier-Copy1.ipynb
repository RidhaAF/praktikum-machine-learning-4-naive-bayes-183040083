```python
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
outlook_enc=pd.get_dummies(X['Outlook'],prefix='Outlook')
```

```python
outlook_enc
```

```python
windy_enc=pd.get_dummies(X['Windy'],prefix='Windy')
```

```python
windy_enc
```

```python
encoded_outlook_windy=pd.concat([outlook_enc,windy_enc],axis=1)
```

```python
encoded_outlook_windy
```

```python
encoded_outlook_windy.drop(['Outlook_overcast','Windy_False'],axis=1,inplace=True)
encoded_outlook_windy = pd.concat([X,encoded_outlook_windy],axis=1)
```

```python
encoded_outlook_windy
```

```python
encoded_outlook_windy.drop(['Outlook','Windy'],axis=1,inplace=True)
```

```python
X
```

```python
encoded_outlook_windy
```

```python
X=encoded_outlook_windy
```

```python
# memisahkan dataset ke dalam set training dan set testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)
```

```python
# melatih model dengan set training
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
# membuat prediksi hasil set testing
predictions = gnb.predict(X_test)
```

```python
# membandingkan hasil prediksi dan label sesungguhnya
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
